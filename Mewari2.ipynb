### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
### have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
### jobs data.
### This task will be done in following steps:
### 1. First get the webpage https://www.shine.com/
### 2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
### location” field.
### 3. Then click the search button.
### 4. Then scrape the data for the first 10 jobs results you get.
### 5. Finally create a dataframe of the scraped data.
#### Note: All of the above steps have to be done in code. No step is to be done manually.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time

#Lets first connect to the driver
driver=webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32")

#Visiting Website
driver.get('https://www.shine.com/')

#Enter “Data Analyst” in “Skill, Designations, Companies” field 
location” field
designation=driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Analyst')

#enter “Bangalore” in “enter the location field"
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Banglore')

#search button
search=driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()

job_title=[]
company_name=[]
experience_required=[]
job_location=[]





```

```python
#Scraping job title from given path
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)

#Scraping job location from given path
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

#Scraping job company's name from given path
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

#Scraping Experience required from given path
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
```

```python
import pandas as pd
df=pd.DataFrame({'Job Title':job_title,'Comapany':company_name,'Location':job_location,'Experience':experience_required})
df
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time

#Lets first connect to the driver
driver=webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32")

#Visiting Website
driver.get('https://www.flipkart.com/')

product=driver.find_element(By.CLASS_NAME,'_3704LK')
product.send_keys('sunglasses')

```

```python
search=driver.find_element(By.CLASS_NAME,'L0Z3Pu')
search.click()
```

```python
Brand=[]
Product_Description=[]
Price=[]
Discount=[]

start=0
end=2
for page in range(start,end):
    Brand_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in Brand_tags:
        title=i.text
        Brand.append(title)

    product_tags=driver.find_elements(By.XPATH,'//div[@class="_2B099V"]')
    for i in product_tags:
        product_dis=i.find_element(By.TAG_NAME,'a').text
        Product_Description.append(product_dis)
    
    Price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in Price_tags:
        title=i.text
        Price.append(title)
    
    Discount_tags=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in Discount_tags:
        dis=i.find_element(By.TAG_NAME,'span').text
        Discount.append(dis)
    next_button=driver.find_element(By.XPATH,"//a[@class='_1LKTO3']")
    next_button.click()
    time.sleep(3)
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time

#Lets first connect to the driver
driver=webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32")

#Visiting Website
driver.get('https://www.azquotes.com/')

#clicking on top quotes
top=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top.click()
```

```python
Quote=[]
Author=[]
Type=[]

#for next pages
start=0
end=9
for page in range(start,end):
    quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tags:
        title=i.text
        Quote.append(title)
    
    author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tags:
        title=i.text
        Author.append(title)

    type_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_tags:
        title=i.text
        Type.append(title)
    next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
    next_button.click()
    time.sleep(2)

quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote_tags:
    title=i.text
    Quote.append(title)
    
author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
for i in author_tags:
    title=i.text
    Author.append(title)

type_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in type_tags:
    title=i.text
    Type.append(title)
```

```python
df=pd.DataFrame({'Quote':Quote,'Author':Author,'Type':Type})
df
```
